<b>Project Overview:-</b>

Electricity plays a vital role in modern economic development, technological advancement, and industrial growth. Over the past few decades, global electricity demand has increased significantly due to rapid population growth, urbanization, industrial expansion, and the growing dependence on digital technologies. As societies become more reliant on electricity for transportation, manufacturing, communication, and daily activities, understanding electricity consumption patterns has become increasingly important.

This project analyzes global electricity consumption trends and energy demand growth from 2000 to 2024. The analysis focuses on examining how electricity generation has evolved across countries and how different energy sources contribute to overall electricity production. By studying electricity generation data from sources such as coal, gas, solar, wind, and hydro, the project aims to identify patterns in energy production and understand the changes occurring in the global energy landscape.

In recent years, energy production has also become closely connected to environmental sustainability and climate change mitigation. Many countries are gradually transitioning from fossil fuel–based electricity generation to renewable energy sources in order to reduce carbon emissions and create more sustainable energy systems. Analyzing electricity generation trends helps highlight these shifts and provides insights into how different regions are responding to the increasing demand for electricity while moving toward cleaner energy solutions.

The dataset used in this analysis was obtained from Our World in Data, which provides publicly available global datasets compiled from multiple reliable international sources. The dataset includes country-level information on electricity generation, population, economic indicators such as GDP, and electricity production from various energy sources. This dataset provides a comprehensive foundation for exploring global electricity consumption patterns and understanding long-term energy trends across different countries and regions.

<b>STEP 2: DATA PROCESSING</b>

<b>1. Sorting Necessary Columns</b>

```python
#Necessary columns

Sorted_data=['country','year','iso_code','electricity_generation','population','gdp',
                  'energy_per_capita','coal_electricity',
                  'gas_electricity','solar_electricity','wind_electricity','hydro_electricity']

df=df[Sorted_data]
```

:- Key necessary columns have been sorted & stored seperately from the dataset for efficient analysis.

```python
df.shape
```

<b>4. Finding Total No. of Null values in the dataset</b>

```python
#Total Missing/Null Values

df.isnull().sum()
```

:- Null values are found in all columns except 'country' & 'year'

<b>5. Filling Null values with Mean Imputation</b>

```python
#Ecnomical indicator null columns filled with mean

df['gdp']=df['gdp'].fillna(df['gdp'].mean())
df['energy_per_capita']=df['energy_per_capita'].fillna(df['energy_per_capita'].mean())
```

:- The null values in GDP & Energy_Per Capita columns were filled with mean imputation since GDP & Energy capita of a country will never be zero.

<b>6. Filling Null values with Zero</b>

```python
#Electricity columns filling with zero

df['coal_electricity']=df['coal_electricity'].fillna(0)
df['gas_electricity']=df['gas_electricity'].fillna(0)
df['solar_electricity']=df['solar_electricity'].fillna(0)
df['wind_electricity']=df['wind_electricity'].fillna(0)
df['hydro_electricity']=df['hydro_electricity'].fillna(0)
```

:- The null values in the the above columns were filled with zero since the countries hasn't produced any energy through that source.

<b>7. Dropping Data</b>

```python
#Removing rows not having essential data

df=df.dropna(subset=['electricity_generation'])
```

:- Electricity generation is a core variable for the analysis, all rows having null values in the column 'electricity_generation' is dropped.

<b>8. Rechecking the for Null Values</b>

```python
df.isnull().sum() 
```

:- Most of the null values were handle except 2 columns ('iso_code' & 'population'), this implies that non-country data has been included in the dataset and it will be filtered out in filtering process.

<b>9. Checking for Duplicates</b>

```python
#Finding duplicates

df.duplicated().sum()
```

:- No duplicates found in the dataset

```python
#Removing Duplicates

df=df.drop_duplicates()
```

:- To ensure data integrity, command for dropping duplicates has been used.

<b>10. Checking Datatypes</b>

```python
#Checking datatypes

df.dtypes
```

:- The datatype are correct for all columns. But for the column 'country' it would be good to change it to a category datatype.

<b>11. Fixing Datatypes</b>

```python
#Fixing the datatypes

df['year']=df['year'].astype(int)
df['country']=df['country'].astype('category')
```

:- The datatype for column 'country' & 'year' has been fixed.

<b>12. Rechecking Datatypes</b>

```python
df.dtypes
```

:- All datatypes has been fixed.

<b>13. Creating Derived Columns</b>

```python
#Creating Derived Columns

#Electricity Per Capita
df['electricity_per_capita']=df['electricity_generation']/df['population']

#Total Renewable Electricity
df['renewable_electricity']=(df['solar_electricity'] + df['wind_electricity'] + df['hydro_electricity'])

#Fossil Electricity
df['fossil_electricity'] = (df['coal_electricity'] + df['gas_electricity'])
```

:- 3 Derived columns has been created
- Electricity Per Capita - Total electricity generation by the population,This helps compare countries of different population sizes.
- Total Renewable Electricity - Adding all renewable source, This helps measure renewable energy growth.
- Fossil Electricity - Adding non-renewable source

<b>14. Filtering Data</b>

```python
#Filtering year from 2000 to 2024
df=df[(df['year'] >= 2000) & (df['year'] <= 2024)]

#Filtering the actual country from regions
df = df[df['iso_code'].str.len() == 3]

df['country'].unique()
```

:- The year in the dataset has been filtered according to subject analysis and all non country regions has been filtered out for better analysis.

```python
#Global electricity generation per year
global_generation=df.groupby('year')['electricity_generation'].sum().reset_index()

#Electricity generation by country
country_generation=df.groupby('country')['electricity_generation'].sum().reset_index()

#Renewable electricity trend
renewable_trend=df.groupby('year')['renewable_electricity'].sum().reset_index()
```

<h5> <b>STEP 3 : EXPLORATORY DATA ANALYSIS & VISUALIZATION<b></h5>

<B>1. Distribution of Electricity Generation </b>

```python
plt.figure(figsize=(10,6))
sns.histplot(df['electricity_generation'], bins=50,color='teal')
plt.title("Distribution of Electricity Generation")
plt.xlabel("Electricity Generation (TWh)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

Insight: Most countries produce low electricity, while a small number of countries generate extremely high levels. This suggests a strong imbalance in global electricity production, where a few major economies dominate energy generation.The presence of outliers highlights the significant gap between developed and developing countries in terms of energy infrastructure and capacity.

<b>2. Energy per Capita Distribution</b>

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['energy_per_capita'],color='orange')
plt.title("Energy Consumption per Capita Distribution", fontsize=14)
plt.xlabel("Energy per Capita (kWh per person)")
plt.ylabel("Energy per Capita (kWh per person)")
plt.grid(True)
plt.show()
```

Insight: The box plot shows a wide variation in energy consumption per capita across countries, with several extreme outliers. Developed countries tend to have significantly higher per capita energy consumption due to better infrastructure, higher standards of living, and increased use of electrical appliances. In contrast, developing countries show lower consumption levels, indicating disparities in access to energy resources and economic development.

<b>3. Electricity vs GDP</b>

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='gdp', y='electricity_generation',color='purple', data=df)
plt.title("Relationship Between GDP and Electricity Generation",fontsize=14)
plt.xlabel("GDP (in USD)")
plt.ylabel("Electricity Generation (TWh)")
plt.grid(True)
plt.show()
```

Insight: The scatter plot reveals a strong positive relationship between GDP and electricity generation. Countries with higher GDP tend to produce and consume more electricity, reflecting their higher industrial activity and economic output. This trend confirms that electricity consumption is closely tied to economic development, although some variations exist due to differences in energy efficiency and policy decisions.

<b>4. Electricity vs Population</b>

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='population', y='electricity_generation',color='teal', data=df)
plt.title("Electricity Generation vs Population")
plt.xlabel("Population (Number of People)")
plt.ylabel("Electricity Generation (TWh)")
plt.grid(True)
plt.show()
```

Insight: The relationship between population and electricity generation is positive but more scattered compared to GDP. While countries with larger populations generally require more electricity, population alone does not fully explain electricity demand. Some highly populated countries have lower electricity generation due to lower industrialization levels, highlighting that economic development plays a more crucial role than population size.

<b>5. Top 10 Countries by Electricity Generation </b>

```python
top10 = df.groupby('country')['electricity_generation'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
top10.plot(kind='bar',color='skyblue')
plt.title("Top 10 Electricity Producing Countries",fontsize=14)
plt.xlabel("Country")
plt.ylabel("Total Electricity Generation (TWh)")
plt.show()
```

Insight: The bar chart clearly shows that a small number of countries dominate global electricity production. Countries such as China, the United States, and India contribute the largest share due to their large populations, strong economies, and extensive industrial sectors. This concentration indicates that global electricity generation is heavily dependent on a few major contributors.

<b>6. Correlation Heatmap</b>

```python
numeric_df = df.select_dtypes(include=['number'])


plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True,cmap="coolwarm")
plt.title("Correlation Between Variables", fontsize=14)
plt.show()
```

Insight: The correlation heatmap highlights strong positive relationships between electricity generation, GDP, and population. Among these, GDP shows the strongest correlation, indicating that economic activity is a key driver of electricity demand. Additionally, energy-related variables such as fossil fuel and renewable electricity also show meaningful relationships, helping to understand the structure of energy production.

<b>7. Global Electricity Trend Over Time</b>

```python
global_trend = df.groupby('year')['electricity_generation'].sum()

plt.figure(figsize=(10,6))
global_trend.plot(color='green',marker='o')
plt.title("Global Electricity Generation Trend (2000–2024)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Electricity Generation (TWh)")
plt.grid(True)
plt.show()
```

Insight: The line chart shows a steady and continuous increase in global electricity generation from 2000 to 2024. This trend reflects the growing demand for electricity driven by industrial growth, urbanization, and technological advancements. The consistent upward pattern indicates that global energy demand is unlikely to decline in the near future.

<b>8. Renewable vs Fossil Energy Trend</b>

```python
energy_trend = df.groupby('year')[['renewable_electricity','fossil_electricity']].sum()

plt.figure(figsize=(10,6))
energy_trend.plot(color=['green','red'],linewidth=2)
plt.title("Renewable vs Fossil Electricity Trend")
plt.xlabel("Year")
plt.ylabel("Electricity Generation (TWh)")
plt.legend(["Renewable", "Fossil"])
plt.grid(True)
plt.show()
```

Insight: The comparison between renewable and fossil electricity shows that while fossil fuels still dominate overall electricity generation, renewable energy sources have been increasing steadily over time. The gap between the two is gradually narrowing, indicating a slow but consistent transition toward cleaner energy sources.

<b>9. Electricity Source Contribution</b>

```python
sources = df[['coal_electricity','gas_electricity','solar_electricity','wind_electricity','hydro_electricity']].sum()

plt.figure(figsize=(7,7))
sources.plot(kind='pie', autopct='%1.1f%%',colors=sns.color_palette('pastel'))
plt.title("Electricity Generation by Source")
plt.show()
```

Insight: The pie chart illustrates the proportion of electricity generated from different energy sources. Fossil fuels such as coal and gas contribute a significant share, while renewable sources like hydro, solar, and wind also make a noticeable contribution. This distribution highlights the current dependency on fossil fuels while also showing the growing presence of renewable energy.

<b>10. Renewable Growth Trend</b>

```python
renewable_trend = df.groupby('year')['renewable_electricity'].sum()

plt.figure(figsize=(10,6))
renewable_trend.plot(color='darkgreen',marker='o')
plt.title("Growth of Renewable Electricity (2000–2024)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Renewable Electricity Generation (TWh)")
plt.grid(True)
plt.show()
```

Insight: The renewable electricity trend shows a significant increase, especially after 2015. This growth reflects global efforts to adopt sustainable energy solutions driven by environmental concerns, technological advancements, and supportive government policies. The rapid rise in renewable energy indicates a strong shift toward cleaner energy systems.

<H5><b>STEP 4 : INSIGHTS</b></H5>

<B>1 — Growth in Global Electricity Demand</B>

The analysis shows a consistent and significant increase in global electricity generation from 2000 to 2024. This upward trend reflects the growing demand for electricity driven by rapid industrialization, urbanization, and the expansion of digital technologies. The increasing reliance on electricity across sectors such as transportation, manufacturing, and communication highlights its critical role in modern economies.

<B>2 — Dominance of Major Economies</B>

A small number of countries, particularly China, the United States, and India, contribute a disproportionately large share of global electricity generation. These countries have large populations, strong industrial bases, and high economic activity, which results in higher electricity demand. This concentration indicates that global electricity production is heavily influenced by a few major economies.

<B>3 — Strong Relationship Between GDP and Electricity Generation</B>

The analysis reveals a strong positive correlation between a country’s GDP and its electricity generation. Countries with higher economic output tend to consume and produce more electricity due to increased industrial activity, infrastructure development, and higher standards of living. This highlights electricity as a key indicator of economic growth and development.

<B>4 — Population Has a Moderate Impact on Electricity Demand</B>

While population size does influence electricity demand, it is not the only determining factor. Some countries with large populations may have relatively lower electricity consumption due to lower levels of industrialization or limited infrastructure. This indicates that economic development and technological advancement play a more significant role than population alone.

<B>5 — Rapid Growth of Renewable Energy Sources</B>

Renewable energy sources such as solar, wind, and hydro have shown significant growth, particularly after 2010. This increase reflects global efforts to transition toward cleaner and more sustainable energy systems. Technological advancements, government policies, and environmental concerns have contributed to the rising adoption of renewable energy.

<B>6 — Continued Dependence on Fossil Fuels</B>

Despite the growth in renewable energy, fossil fuels such as coal and natural gas continue to dominate global electricity generation. Many countries still rely heavily on these sources due to existing infrastructure and cost considerations. This indicates that the transition to renewable energy is ongoing but not yet complete.

<B>7 — Large Variation in Electricity Consumption per Capita</B>

There is a significant variation in electricity consumption per capita across different countries. Developed countries tend to have higher electricity consumption per person due to better infrastructure, higher living standards, and greater access to electrical appliances and services. In contrast, developing countries show lower consumption levels.

<B>8 — Transition Toward a Diversified Energy Mix</B>

The analysis indicates a gradual shift in the global energy mix, with countries increasingly incorporating renewable energy sources alongside traditional fossil fuels. This diversification helps improve energy security and reduces dependence on a single source of energy, contributing to more stable and sustainable energy systems.

<B>9 — Faster Growth in Emerging Economies</B>

Emerging economies, particularly in Asia, are experiencing faster growth in electricity demand compared to developed nations. This is mainly due to rapid industrialization, urban expansion, and increasing access to electricity. These countries are expected to play a major role in shaping future global energy demand.

<B>10 — Acceleration of Energy Transition After 2015</B>

The data shows a noticeable acceleration in renewable energy adoption after 2015. This trend can be linked to global climate agreements, policy changes, and declining costs of renewable technologies. As a result, renewable electricity generation has increased significantly, indicating a stronger global commitment toward sustainable energy.
